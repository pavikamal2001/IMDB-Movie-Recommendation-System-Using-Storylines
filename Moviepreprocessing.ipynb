```python
import re
import nltk
import contractions
import pandas as pd


from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download required NLTK resources (run once)
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Initialize stopwords & lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()


def text_preprocessing(text):
    text = str(text).lower()
    text = contractions.fix(text)
    text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation & numbers

    tokens = word_tokenize(text)

    processed_text = ' '.join(
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in stop_words
    )

    return processed_text
```

```python
df=pd.read_csv('/content/imdb_movies_2024.csv')
df
```

```python
df.columns = df.columns.str.strip()

print(df.columns)
```

```python
df = df.dropna(subset=["Story_line"]).reset_index(drop=True)
```

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    text = str(text).lower()                 # lowercase
    text = contractions.fix(text)            # expand contractions
    text = re.sub(r"[^a-z\s]", "", text)     # remove punctuation & numbers
    text = re.sub(r"\s+", " ", text).strip()
    text=  ' '.join([word  for word in text.split() if word.isalpha()]) # remove extra spaces
    return text
```

```python
df["clean_story"] = df["Story_line"].apply(clean_text)
df[["Story_line", "clean_story"]]
```

```python
df['Cleaned_Text']=df['clean_story'].apply(text_preprocessing)
df
```

```python
stop_words=set(stopwords.words('english'))
stop_words
lemmatizer=WordNetLemmatizer()
lemmatizer
```

```python
## Tokenization
df["tokens"] = df["clean_story"].apply(word_tokenize)
df[["clean_story", "tokens"]].head()
```

```python
##Remove Stopwords
```

```python
df["tokens_no_stopwords"] = df["tokens"].apply(
    lambda words: [w for w in words if w not in stop_words]
)

df[["tokens", "tokens_no_stopwords"]].head()
```

```python
##Lemmatization
```

```python
df["tokens_lemmatized"] = df["tokens_no_stopwords"].apply(
    lambda words: [lemmatizer.lemmatize(w) for w in words]
)

df[["tokens_no_stopwords", "tokens_lemmatized"]].head()
```

```python
df["final_clean_storyline"] = df["tokens_lemmatized"].apply(lambda x: " ".join(x))

df[["Title", "Story_line", "final_clean_storyline"]].head()
```

```python
df["Title_clean"] = df["Title"].str.replace(r"^\d+\.\s*", "", regex=True)
df[["Title", "Title_clean"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=5000)

# Convert storylines into numerical vectors
tfidf_matrix = tfidf.fit_transform(df["final_clean_storyline"])

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
tfidf.get_feature_names_out()[:20]
```

```python
##cosine similiarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

print(cosine_sim_matrix.shape)
```

```python
def recommend_movies(movie_title, top_n=5):
    movie_title = movie_title.lower().strip()

    match = df[df["Title_clean"].str.lower() == movie_title]

    if match.empty:
        return f"Movie '{movie_title}' not found!"

    idx = match.index[0]

    sim_scores = list(enumerate(cosine_sim_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:top_n+1]
    movie_indices = [i[0] for i in sim_scores]

    result = df.iloc[movie_indices][["Title_clean", "Story_line"]].copy()
    result["Similarity Score"] = [s[1] for s in sim_scores]

    return result
```

```python
recommend_movies("The Substance", top_n=5)
```

```python
def recommend_by_storyline(user_storyline, top_n=5):
    # preprocess input storyline
    user_clean = text_preprocessing(user_storyline)

    # convert to tfidf vector
    user_vec = tfidf.transform([user_clean])

    # similarity with all movies
    sim_scores = cosine_similarity(user_vec, tfidf_matrix).flatten()

    # top n results
    top_indices = sim_scores.argsort()[-top_n:][::-1]

    result = df.iloc[top_indices][["Title_clean", "Story_line"]].copy()
    result["Similarity Score"] = sim_scores[top_indices]

    return result
```

```python
recommend_by_storyline(
    "A young wizard joins a magical school and fights dark forces",
    top_n=5
)
```
